```python
import pandas as pd
from scipy.stats import zscore

# Salary data
df = pd.DataFrame({
    'Employee_ID': ['E01','E02','E03','E04','E05','E06','E07'],
    'Salary': [48000, 52000, 50500, 49200, 51000, 120000, 18000]
})

# Z-score calculation
df['Z_Score'] = zscore(df['Salary'])
print(df['Z_Score'])

# Detect outliers
outliers = df[abs(df['Z_Score']) > 2]
print(outliers)
```
